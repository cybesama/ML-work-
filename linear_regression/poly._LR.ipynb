```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= 4 * np.random.rand(100,1)+3
y=2.5+2.3*x+4*x*x+np.random.randn(100,1)
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
plt.scatter(x_train,y_train)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=True,)
x_train_normalised=poly.fit_transform(x_train)
x_test_normalised= poly.transform(x_test)
x_train_normalised
```

```python
from sklearn.linear_model import LinearRegression 
lr=LinearRegression()
lr.fit(x_train_normalised,y_train)
y_predict=lr.predict(x_test_normalised)
```

```python
from sklearn.metrics import r2_score as r2
score=r2(y_test,y_predict)
print(score)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
plt.scatter(x_test,y_test,color='r')
plt.scatter(x_test,y_predict,color='g')
```
