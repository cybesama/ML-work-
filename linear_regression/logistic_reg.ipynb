```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
x,y=make_classification(n_classes=2,random_state=42,n_samples=1000,n_features=10)
```

```python
pd.DataFrame(x,y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
```

```python
x_train.shape
y_train.shape
```

```python
lg.fit(x_train,y_train,)
```

```python
y_predict=lg.predict(x_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report 
```

```python
print(accuracy_score(y_test,y_predict))
print(confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
```

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import StratifiedKFold
cvs=StratifiedKFold()
```

```python
# hyper paameter tuning (using gridsearchCV)
penalty=['l1','l2','elasticnet']
c_values=[.01,100,10,1,.1]
solver=['newton-cg','lbfgs','liblinear','sag','saga']

params=dict(penalty=penalty,C=c_values,solver=solver)

gs=GridSearchCV(estimator=lg,param_grid=params,scoring='accuracy',n_jobs=-1,cv=cvs)
gs
result=gs.fit(x_train,y_train)
```

```python
print(result.best_score_)
print(result.best_params_)
```

```python
y_predict_new=gs.predict(x_test)
```

```python
print(accuracy_score(y_test,y_predict_new))
print(confusion_matrix(y_test,y_predict_new))
print(classification_report(y_test,y_predict_new))
```

```python
from sklearn.datasets import make_classification
x1,y1=make_classification(n_classes=3,random_state=42,n_samples=1000,n_informative=3,n_features=10)
```

```python
pd.DataFrame(x1,y1)
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lg1=LogisticRegression(multi_class='ovr')
lg1.fit(x1_train,y1_train)
y1_predict=lg1.predict(x1_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report 
print(accuracy_score(y1_test,y1_predict))
print(confusion_matrix(y1_test,y1_predict))
print(classification_report(y1_test,y1_predict))
```
