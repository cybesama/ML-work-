```python
import seaborn as sns
df=sns.load_dataset('tips')
```

```python
x=df[['day','size','sex','time','smoker']]
y=df['total_bill']
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder

le2=LabelEncoder()
le3=LabelEncoder()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,train_size=.2,random_state=10)
```

```python

X_train['smoker']=le2.fit_transform(X_train['smoker'])
X_train['time']=le3.fit_transform(X_train['time'])
```

```python

X_test['smoker']=le2.transform(X_test['smoker'])
X_test['time']=le3.transform(X_test['time'])
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('Onehot',OneHotEncoder(drop='first'),[0])],remainder='passthrough')
```

```python
X_train=ct.fit_transform(X_train)
X_train
```

```python
X_test=ct.transform(X_test)
```

```python
import numpy as np
x2=np.hstack((x11,x22))
x2
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
```

```python
from sklearn.model_selection import train_test_split
X2_train,X2_test,Y2_train,Y2_test=train_test_split(x2,y2,train_size=.2,random_state=10)
```

```python
gnb.fit(X2_train,Y2_train)
```

```python
y2_pred=gnb.predict(X2_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(confusion_matrix(y2_pred,Y2_test))
print(accuracy_score(y2_pred,Y2_test))
print(classification_report(y2_pred,Y2_test))
```
